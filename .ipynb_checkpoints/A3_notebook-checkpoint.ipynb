## Plotting raw data

```python
import matplotlib.pyplot as plt
import pandas as pd

FEATURE_SIZE = 150
TRAINING_SIZE = 3600

def make_plot(filename):
    df = pd.read_csv(filename)
    plt.plot(df['relative_time'], df['AccX'], label="X Axis")
    plt.plot(df['relative_time'], df['AccY'], label="Y Axis")
    plt.plot(df['relative_time'], df['AccZ'], label="Z Axis")
    plt.plot(df['relative_time'], df['DecibelSource'], label="Decibels")
    plt.legend()
    plt.title(filename[:-4])
    plt.xlabel('Relative Time Stamp')
    plt.show()
    return df
```

```python
web_df = make_plot('webBrowsing.csv')
```

```python
run_df = make_plot('run.csv')
```

```python
walk_df = make_plot('walk.csv')
```

```python
stairs_df = make_plot('stairs.csv')
```

```python
drive_df = make_plot('driving.csv')
```

## Make features

```python
def make_features(df, label):
    res = {}
    res['x_train'] = []
    res['x_test'] = []
    start, end = 0, FEATURE_SIZE
    
    while end <= len(df):
        if end <= TRAINING_SIZE:
            train_test_key = '_train'
        else:
            res['y' + train_test_key] = [label for i in range(len(res['x' + train_test_key]))]
            train_test_key = '_test'
        data = df.iloc[start : end]
        
        # Using mean, SD, and .25, .5, .75 quantiles for feature values
        mean_vals = data.mean().tolist()
        sd_vals = data.std().tolist()
        quant25 = data.quantile(.25).tolist()
        med = data.quantile(.5).tolist()
        quant75 = data.quantile(.75).tolist()
        
        res['x' + train_test_key].append(mean_vals)
        res['x' + train_test_key].append(sd_vals)
        res['x' + train_test_key].append(quant25)
        res['x' + train_test_key].append(med)
        res['x' + train_test_key].append(quant75)
        
        start, end = start+150, end+150
    res['y' + train_test_key] = [label for i in range(len(res['x' + train_test_key]))]
    return res

web_dict    = make_features(web_df, 'web')
run_dict    = make_features(run_df, 'run')
walk_dict   = make_features(walk_df, 'walk')
stairs_dict = make_features(stairs_df, 'stairs')
drive_dict  = make_features(drive_df, 'drive')
```

```python
# Combine into one array
x_train = web_dict['x_train'] + run_dict['x_train'] + walk_dict['x_train'] + stairs_dict['x_train'] + drive_dict['x_train']
y_train = web_dict['y_train'] + run_dict['y_train'] + walk_dict['y_train'] + stairs_dict['y_train'] + drive_dict['y_train']

x_test = web_dict['x_test'] + run_dict['x_test'] + walk_dict['x_test'] + stairs_dict['x_test'] + drive_dict['x_test']
y_test = web_dict['y_test'] + run_dict['y_test'] + walk_dict['y_test'] + stairs_dict['y_test'] + drive_dict['y_test']
```

## Decision tree predictions

```python
from sklearn import tree

# Use one hot encoding for categorical y labels
one_hot_enc_train = pd.get_dummies(y_train)
one_hot_enc_test = pd.get_dummies(y_test)

clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train, one_hot_enc_train)
y_pred = clf.predict(x_test)

score = clf.score(one_hot_enc_test, y_pred)
score
```

```python
labels = one_hot_enc_train.columns

y_prob_pred = clf.predict_proba(x_test)
probs = []
for arr in y_prob_pred:
    row_sum = 0
    for row in arr:
        row_sum += row[0]
    probs.append(row_sum / float(len(arr)))

for i in range(len(labels)):
    print(f'{labels[i]} --> {probs[i]}')
```
